## 🎓 Distribution Analysis of Student Scores

### 📘 Math Score
The math scores range from 0 to 100, with most students scoring between 60 and 80. The distribution is slightly skewed to the left, meaning more students performed well in math. A small number of students scored very low, indicating possible outliers. Overall, the majority of students had average to above-average performance in math.

### 📗 Reading Score
Reading scores are fairly normally distributed, with a peak around 70 to 80. Most students performed well in reading, and there are fewer very low or very high scores compared to math. The distribution is more balanced, suggesting that reading is a strength for many students.

### 📕 Writing Score
The writing score distribution also follows a near-normal pattern, though slightly skewed to the left. Like reading, most students scored between 65 and 80. However, a few students scored below 40, which might indicate difficulties with writing or other influencing factors. In general, writing performance is closely aligned with reading scores.

### ✍️ Summary
Across all three subjects, students tend to perform better in reading and writing than in math. There are some outliers in each subject, but the majority of scores cluster in the average-to-good range (60–80).


```python
sns.histplot(x='math score',kde=True,data=df)
plt.title("Distribution of Maths score")
plt.show()
```

```python
sns.histplot(x='writing score',kde=True,data=df,color="cyan")
plt.title("Distribution of Writing score")
plt.show()
```

```python
sns.histplot(x='reading score',kde=True,data=df,color='black')
plt.title("Distribution of Reading score")
plt.show()
```

## 🍱 Lunch Type vs Student Scores

We explore whether lunch type (an indicator of socioeconomic status) affects academic performance.

### 🔍 Boxplot Insights:
- Shows distribution of math, reading, and writing scores by lunch type.
- Use to observe:
  - Median score difference
  - Variability of scores
  - Presence of outliers


#### 📈 Expectation:
If students with **standard lunch** consistently score higher, it may suggest that **nutrition and economic factors influence learning outcomes**.


```python
sns.boxplot(x='lunch', y='math score', data=df)
plt.title('Math Score vs Lunch Type')
plt.show()
```

```python
sns.boxplot(x='lunch', y='writing score', data=df,color='green')
plt.title('Math Score vs Lunch Type')
plt.show()
```

```python
sns.boxplot(x='lunch', y='reading score', data=df,color="red")
plt.title('Math Score vs Lunch Type')
plt.show()
```

### 🌡️ Heatmap Insights:
- Displays the **average score** in each subject for each lunch group.
- Useful to quickly compare overall performance between groups.

```python
lunch_avg_scores = df.groupby('lunch')[['math score', 'reading score', 'writing score']].mean()
plt.figure(figsize=(8, 4))
sns.heatmap(lunch_avg_scores, annot=True, cmap='YlGnBu', fmt='.1f')
plt.title('Average Scores by Lunch Type')
plt.xlabel('Subject')
plt.ylabel('Lunch Type')
plt.show()
```

## 📊 Gender vs Parental Education

In this section, we explore the relationship between **gender** and **parental education level**. Specifically, we are interested in seeing:
- How many **males** and **females** fall into each **parental education category**.
- Whether there is any noticeable **gender imbalance** at different levels of parental education.

### 🔍 Why This Analysis?

- It helps us understand if one gender is more likely to come from a certain **educational background** of their parents.
- This could provide insights into **gender-based educational support** and how it might impact academic performance.

### 📊 Countplot Insights

A **countplot** is used to show the **distribution** of genders across different parental education levels:
- **X-axis**: Parental education levels (e.g., High School, Some College, etc.).
- **Hue**: Gender (Males and Females).
- **Y-axis**: Number of students in each category

```python
sns.countplot(x='parental level of education', hue='gender', data=df,color='orange')
plt.title('Gender vs Parental Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
